```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def inital_analysis(filen):
  df=pd.read_csv(filen)
  print(df.head())
  print(df.info())
  print(df.describe())
  print(df.isnull().sum())
  print(df.duplicated().sum())
  return df
df=inital_analysis('productivity.csv')

```

```python
def handlemissingval(df):
  df=df.drop('wpl')
```

```python
def univariateanalusis(df,target):
  plt.figure(figsize=(10,10))
  sns.histplot(df[target],kde=True,bins=30)
  plt.show()
for i in df.select_dtypes(include='number').columns:
  univariateanalusis(df,i)

def catunivariateanalysis(df,target):
  plt.figure(figsize=(10,10))
  sns.countplot(df[target])
  plt.show()
for i in df.select_dtypes(include='object').columns:
  catunivariateanalysis(df,i)
```

```python
def bivariateanalysis(df,y):
  for x in df.select_dtypes(include='number').columns:
    plt.figure(figsize=(10,10))
    sns.lineplot(x=df[x],y=df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
bivariateanalysis(df,'actual_productivity')
def catbivariateanalysis(df,y):
  for x in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(10,10))
    sns.barplot(x=df[x],y=df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
catbivariateanalysis(df,'actual_productivity')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True)
plt.show()
```

```python
class handle_missing_values():
  def __init__(self,df):
    self.df=df
  def meanmethod(self,col_name):
    self.df[col_name].fillna(self.df[col_name].mean(),inplace=True)
  def modemethod(self,col_name):
    self.df[col_name].fillna(self.df[col_name].mode()[0],inplace=True)
  def medianmethod(self,col_name):
    self.df[col_name].fillna(self.df[col_name].median(),inplace=True)
  def value_handling(self,col_name,value):
    self.df[col_name]=self.df[col_name].fillna(value)
```

```python
miss=handle_missing_values(df1)
miss.value_handling('Medal','No Medal')
miss.meanmethod('Age')
miss.medianmethod('Height')
miss.meanmethod('Weight')
print(df1.isnull().sum())
```

```python
miss2=handle_missing_values(df2)
miss2.value_handling('notes','No Notes')
miss2.value_handling('region','Germany')
print(df2.isnull().sum())
```
